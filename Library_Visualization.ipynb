```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv('archive/Library_Usage.csv')
df = df[:2000]
df = df[df['Home Library Definition'] != None]
```

```python
text_style = {
    "labelPadding": 5,
    "labelColor": "#635F5D",
    "labelFontStyle": "bold",
    "labelFont": "Ubuntu Mono",
    "labelFontSize": 12,
    "titlePadding": 22,
    "titleFont": "Ubuntu Mono",
    "titleColor": "#635F5D"
}
axis_style = {
    "gridWidth": 1,
    "gridColor": "#FFFFFF", 
    "domain": False,
    "ticks": False,
    **text_style
}
configure_style = {
    "padding": 22,
    "background": "#EFECEA"
}
title_config = {
    "offset": 22,
    "anchor": "start",
    "fontSize": 20,
    "color": "#635F5D",
    "font": "Ubuntu Mono",
    "subtitleColor": "#8E8883",
    "subtitleFontSize": 12,
    "subtitleFontStyle": "italic",
    "subtitleFont": "Ubuntu Mono",
}
```

```python
lines = alt.Chart(df).mark_line().encode(
        x=alt.X(
        "Total Checkouts:Q",
        axis=alt.Axis(
            labels=False, 
            title="Total Checkouts",
            **axis_style
        ),
        scale=alt.Scale(domain=[0, 7000], nice=False)
    ),
    y = alt.Y('Home Library Definition:N', axis=alt.Axis(
            **axis_style
        ), sort=alt.Sort(field='investment', op='max', order='descending')))

points = alt.Chart(df).mark_circle(size = 100).encode(
    x = alt.X('Total Checkouts:Q'),
    y = alt.Y('Home Library Definition:N', sort=alt.Sort(field='Total Checkouts', op='max', order='descending')),
    color  = alt.Color('Year Patron Registered:N', scale=alt.Scale(scheme='darkblue'))
)

alt.layer(points).configure(**configure_style).properties(title=alt.TitleParams(text="Total checkouts by libraries"))
```

```python
dropdown = alt.binding_select(options = df['Home Library Definition'].unique())
select_country = alt.selection_single(empty = 'none', bind = dropdown, fields = ['Home Library Definition'], init={'Home Library Definition': 'Potrero'})

alt.Chart(df).mark_line().encode(
    x = alt.X(field = 'Year Patron Registered', type = 'quantitative'),
    y = alt.Y(field = 'Total Checkouts', type = 'quantitative'),
    detail = alt.Detail(field = 'Home Library Definition', type = 'nominal'),
    tooltip = alt.Tooltip(field = 'Home Library Definition', type = 'nominal')
).configure(**configure_style).properties(title=alt.TitleParams(text="Total checkouts by year")).add_selection(select_country).transform_filter(select_country)
```

```python
alt.Chart(df).mark_bar().encode(
    x = alt.X(field = 'Age Range', type = 'nominal'),
    y = alt.Y(field = 'Total Checkouts', type = 'quantitative'),
    color = alt.Color(field = 'Age Range', type = 'nominal')
).configure(**configure_style).properties(title=alt.TitleParams(text="Checkouts number by reader's age"))
```
